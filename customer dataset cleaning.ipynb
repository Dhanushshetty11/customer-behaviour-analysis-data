```python
data = pd.read_csv('customer_shopping_behavior.csv')
data.head()
```

```python
# describing the data

data.describe(include = 'all')
```

```python
# checking null values
data.isnull().sum()
```

```python
# now converting the null values
data['Review Rating'] = data.groupby('Category')['Review Rating'].transform(lambda a : a.fillna(a.mean()))
data.isnull().sum()
```

```python

# convert to lower case
data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(' ', '_')
data.columns = data.columns.str.replace('purchase_amount_(usd)', 'purchase_amount')
data.head()
```

```python
# creating new coloumns
label = ["junior" , "senior" , "middle aged" , "old aged"]
data["age_group"] = pd.qcut(data['age'] , q = 4 , labels = label)
data[['age_group', 'age']].head()
```

```python
# creatin new coloumns
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quaterly' : 90,
    'Annually' : 365,
    'Every 3 Months': 90
}

data['purchase_frequency_days'] = data['frequency_of_purchases'].map(frequency_mapping)
data.head()
```

```python
# removing the coloumns
data = data.drop('promo_code_used' , axis=1)
data.columns
```

```python
query = "SELECT * FROM customer ;"
data = pd.read_sql(query, engine)
print(data)



query = "SELECT gender , sum(purchase_amount) FROM customer GROUP BY gender;"
data = pd.read_sql(query, engine)
print(data)
```

```python
# avg more than purchace even after using discount

query = "SELECT  customer_id , purchase_amount FROM customer WHERE discount_applied = 'Yes' AND purchase_amount > (SELECT AVG(purchase_amount) FROM customer);"
data = pd.read_sql(query, engine)
print(data)
```

```python
# top 5 product with rating
query = "SELECT item_purchased , AVG(review_rating) from customer ORDERBY (review_rating) DESC LIMIT 5;"
data = pd.read_sql(query, engine)
print(data)
```

```python
# AVG purchase amount between shipping between standard and express
query = "SELECT AVG(purchase_amount) , shipping_type FROM customer WHERE shipping_type IN ('Standard' , 'Express') GROUP BY shipping_type;"
data = pd.read_sql(query, engine)
print(data)
```

```python
# compare average revenue and total spend between subscribers and non subscribers
query = "SELECT subscription_status , COUNT(customer_id ) , AVG(purchase_amount) AS avgspent , SUM(purchase_amount) AS totalspent FROM customer GROUP BY subscription_status ORDER BY totalspent DESC ;"
data = pd.read_sql(query, engine)
print(data)
```

```python
# customer are repeat buyers for more than 5 time have subscribtions
query = "SELECT COUNT(customer_id) , subscription_status FROM customer WHERE previous_purchases > 5 GROUP BY subscription_status;"
data = pd.read_sql(query, engine)
print(data)
```

```python
# total revenue of each age group
query = "SELECT SUM(purchase_amount) , age_group FROM customer GROUP BY age_group ORDER BY SUM(purchase_amount) DESC;"
data = pd.read_sql(query, engine)
print(data)
```
